<h4>problem1(b)</h4>

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=7, random_state=0).fit(df_norm)
```

```python
clusters=pd.DataFrame(kmeans.labels_, columns=["Cluster"])
clusters
```

```python
clusters = pd.DataFrame(kmeans.labels_, columns=["cluster"])
clusters
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=header)
centroids
```

```python
def cluster_sizes(clusters):
    #clusters is an array of cluster labels for each instance in the data
    size = {}
    cluster_labels = np.unique(clusters)
    n_clusters = cluster_labels.shape[0]

    for c in cluster_labels:
        size[c] = len(clusters[clusters == c])
    return size
```

```python
size = cluster_sizes(np.array(clusters))
size
```

```python
from sklearn import metrics
```

```python
silhouettes = metrics.silhouette_samples(df_norm, np.array(clusters).flatten())
print(silhouettes)
print()
print('silhouettes mean is ',silhouettes.mean())
```

```python
from sklearn.metrics import completeness_score, homogeneity_score
```

```python
np.array(clusters)
```

```python
print("Completeness_score is ",completeness_score(np.array(label)[:,0],np.array(clusters).flatten()))
print("homogeneity_score is ",homogeneity_score(np.array(label)[:,0],np.array(clusters).flatten()))
```
